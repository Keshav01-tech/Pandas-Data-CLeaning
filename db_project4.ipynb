```python
df = df.sort_values(by="Cause", ascending=False)
```

```python
print(df.dtypes)

```

```python
df = df.sort_values(by="Weather Condition", ascending=True) 
```

```python
df_grouped = df.groupby(["Weather Condition", "Cause" ,"Road Condition"]).size().reset_index(name="Count")
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12 ,6)
sns.barplot(
    x="Weather Condition", 
    y="Count", 
    hue="Road Condition", 
    data=df_grouped, 
    palette=['blue', 'orange', 'green', 'yellow', 'red', 'purple']
)

plt.xticks(rotation=45)  # Rotate labels for better readability
plt.show()


```
